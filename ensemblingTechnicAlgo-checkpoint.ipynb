```python
#import libraries : 
import numpy as np
import pandas as pd
from sklearn import metrics
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree 

#from os import system #for other OS (mac/linux)
```

```python
# For the cols regression data was expected - we are getting data in range -> convert such datainto ordinal or categorical data
```

```python
for feature in creditData.columns : #looping across all the cols of the datasets
    if creditData[feature].dtype == 'object': #check for the cols having obj datatype
        creditData[feature] = pd.Categorical(creditData[feature]) #replace the obj col to cat col
```

```python
print(creditData.checking_balance.value_counts())
print(creditData.credit_history.value_counts())
print(creditData.purpose.value_counts())
print(creditData.savings_balance.value_counts())
print(creditData.employment_duration.value_counts())
print(creditData.other_credit.value_counts())
print(creditData.housing.value_counts())
print(creditData.job.value_counts())
print(creditData.phone.value_counts())
```

```python
replaceStruct = {
                "checking_balance":     {"< 0 DM": 1, "1 - 200 DM": 2 ,"> 200 DM": 3 ,"unknown":-1},
                "credit_history": {"critical": 1, "poor":2 , "good": 3, "very good": 4,"perfect": 5},
                 "savings_balance": {"< 100 DM": 1, "100 - 500 DM":2 , "500 - 1000 DM": 3, "> 1000 DM": 4,"unknown": -1},
                 "employment_duration":     {"unemployed": 1, "< 1 year": 2 ,"1 - 4 years": 3 ,"4 - 7 years": 4 ,"> 7 years": 5},
                 "phone":     {"no": 1, "yes": 2 },
                "job":     {"unemployed": 1, "unskilled": 2, "skilled": 3, "management": 4 },
                "default":     {"no": 0, "yes": 1 } 
                    }
```

```python
#those feature with NO cols in their categories
oneHotCols=["purpose","housing","other_credit","job"]
```

```python
print(creditData.employment_duration.value_counts())
```

```python
creditData = creditData.replace(replaceStruct)
creditData
```

```python
creditData.shape
```

```python
creditData.info()
```

```python
creditData=pd.get_dummies(creditData, columns=oneHotCols)
```

```python
creditData.info()
```

```python
creditData.sample(20)
```

```python
#split the data
x = creditData.drop('default' , axis =1)
y = creditData.pop('default')
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=.30, random_state=1)
```

Prunning - reducing overfitting in decision tree(Regularization)

```python
# prepare decision tree model
dTree = DecisionTreeClassifier(criterion='gini' ,max_depth=3 ,random_state=13)
dTree.fit(x_train, y_train)
```

```python
#score of decision tree
print(dTree.score(x_train,y_train)*100)
print(dTree.score(x_test,y_test)*100)
```

```python
#Ensemble learning - Bagging
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
bagCl_model = BaggingClassifier(base_estimator=dTree,n_estimators=50, random_state=1)
```

```python
bagCl_model.fit(x_train,y_train)
```

```python
y_predict = bagCl_model.predict(x_test)
print(bagCl_model.score(x_test,y_test))
```

```python
cm  = metrics.confusion_matrix(y_test, y_predict, labels=[0,1])
df_cm = pd.DataFrame(cm , index=[i for i in ["No" , "Yes"]],  columns=[i for i in ["Predicted No" , "Predicted Yes"]])
plt.figure(figsize=(7,5))
sns.heatmap(df_cm , annot=True, fmt='g')
```
