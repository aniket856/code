```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.cluster import KMeans
from scipy.stats import zscore
```

```python
#load the dataset
tech_support_df = pd.read_csv('E:/pythonWorkspace/excelFiles/technical_support_data.csv')
```

```python
techSupportAttribute = tech_support_df.iloc[:,1:]
```

```python
techSupportAttribute
```

```python
from scipy.stats import zscore
```

```python
techSupportScaled = techSupportAttribute.apply(zscore) #it's optional (just to decrese value)
```

```python
#kde plot 
```

```python
#group the data into similar clusters
#using k-means clustering to group the data based on values. first, it's required to determine the number of groups.
#Elow method to find k value
```

```python
from sklearn.cluster import KMeans
```

```python
#The elbow
from scipy.spatial .distance import cdist #for computiing distance between each pair of points
clusters = range(1,10) #taking range of k values
meanDistrotion=[]
for k in clusters:             
    model=KMeans(n_clusters=k)   # creating models with different k values
    
    model.fit(techSupportScaled)  
    prediction=model.predict(techSupportScaled)

    meanDistrotion.append(sum(np.min(cdist(techSupportScaled, model.cluster_centers_, 'euclidean'), axis=1))/techSupportScaled.shape[0])
```

```python
meanDistortion = meanDistrotion
```

```python
meanDistortion
```

```python
plt.plot(clusters, meanDistrotion, 'bx-')
plt.xlabel('k-value')
plt.ylabel('Mean Distortion Value')
plt.title('Select K value using the elbow method')
```

```python
#Taking k = 3
```

```python
final_model = KMeans(3)
final_model.fit(techSupportScaled)
prediction = final_model.predict(techSupportScaled)
```

```python
prediction
```

```python
# Appending the clusterValue
tech_support_df['cluster_label'] = prediction
```
