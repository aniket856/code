```python
#standarize the data - zscore
from scipy.stats import zscore
```

```python
XstdValue = x.apply(zscore) #standarized the data to make it clear  and look good
XstdValue
```

```python
#convert in dataframe
XstdValue_df = pd.DataFrame(XstdValue)
XstdValue_df
```

```python
#now, will build KNN model
```

```python
# prepare train and test data
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(XstdValue, y, test_size=0.3,random_state=1)
```

```python
#create KNN model
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNN = KNeighborsClassifier(n_neighbors=5,weights='distance')
```

```python
# weights - distance,uniform,
model = KNeighborsClassifier()  
```

```python
model.fit(x_train,y_train)
```

```python
y_predict = model.predict(x_test)
y_predict
```

```python
model.score(x_test,y_predict)*100 
```

```python
from sklearn import metrics
```

```python
conf_m = metrics.confusion_matrix(y_test,y_predict,labels=['M','B'])
```

```python

df_conf_m = pd.DataFrame(conf_m , index = [i for i in ['M' , 'B']] , columns=[i for i in ['Predicted M value' , 'Predicted B value']] )
```

```python
sns.heatmap(df_conf_m , annot=True)
```

```python
from sklearn.model_selection import cross_val_score

all_scores = []

for i in range(1,50):
    KNN_2 = KNeighborsClassifier(n_neighbors=i)
    score = cross_val_score(KNN_2,XstdValue_df, y, cv = 10 )
    all_scores.append(score.mean())
```

```python
all_scores
```

```python
len(all_scores)
```

```python
plt.figure(figsize=(20,12))
plt.plot(range(1,50), all_scores , color="blue" , linestyle='dashed', marker='o' , markerfacecolor="red" , markersize= 15)
plt.title('Accuracy Score vs K-Value')
plt.xlabel('K-value')
plt.ylabel('Accuracy score')
```
